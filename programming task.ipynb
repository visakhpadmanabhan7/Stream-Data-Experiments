```python
import os
import json
import gzip
import pandas as pd
from urllib.request import urlopen
```

```python
# convert list into pandas dataframe
df = pd.read_csv("./ratings_Electronics (1).csv",names=['user_id', 'product_id', 'rating', 'timestamp'], header=None)
df
```

```python
df["date"]=pd.to_datetime(df["timestamp"], unit='s')
df
```

Sorting using date.

```python
sorted_frame = df.sort_values(by='date', ascending=True)
sorted_frame.reset_index(drop=True)
```

```python
sorted_frame=sorted_frame.set_index('date')
```

```python
grouped=sorted_frame.groupby(pd.Grouper(freq='3M')) 
for name,group in grouped:
   print("Group Key->",name)
   print(group)
```

```python
print("Sliding Window")
def my_func(x):
    if(len(x)>0):
        unique_users=x["user_id"].unique()
        print(f"From {x.index[0]} to {x.index[-1]}\n")
        print(f"Number of user reviews in this window {len(x)}")
        print(f"Number of unique user reviews in this window {len(unique_users)}")
        print("\n")
grouped.apply(my_func)
```
